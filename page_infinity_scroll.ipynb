# import needed packages

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from tqdm import tqdm
```

# get the browser

```python
link = "https://stockmages.netlify.app"
browser = webdriver.Edge()
browser.get(link)
```

## script for infinit scroll

```python
for i in tqdm(range(0,1000000,1000)):
    browser.execute_script("window.scrollTo(0,"+ str(i) + ")")
    time.sleep(.1)
```

# get html parser

```python
soup = BeautifulSoup(browser.page_source , "html.parser")
soup
```

# scrap the data

```python
data = []
for sp in soup.find_all("div" , class_ = "container"):
    if('gif' not in sp.find('img').get('src')):
        
        img_link = sp.find("img").get("src")
        tags = sp.find("div" , class_ = "tags").find("span" , class_ = "tag-color").text[7:]
        likes = int(sp.find("div" ,class_ = "likes-comments").find_all("span")[0].text[:-5])
        comments = int(sp.find("div" ,class_ = "likes-comments").find_all("span")[1].text[:-9])

        data.append([img_link,tags,likes,comments])
```

# read the data

```python
data
```

# create dataframe

```python
df = pd.DataFrame(data , columns = ["img_link" , "tags" , "likes" , "comments"])
```

# save the data

```python
df.to_csv("imgs_details.csv")
```
